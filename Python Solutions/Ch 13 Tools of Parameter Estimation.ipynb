# Problem 1
Using the code example for plotting the PDF on page 127, plot the CDF and quantile functions.
# Solution 1

```python
import numpy as np
from scipy.stats import norm
import plotly.express as px

x = norm.ppf(np.arange(0,1,.0001))
y = norm.cdf(x)
fig = px.line(x=x,y=y)
fig.update_layout(title='Normal Distribution CDF')
fig.show()


fig = px.line(x=y,y=x)
fig.update_layout(title='Normal Distribution Percentile')
fig.show()
```

# Problem 2
Returning to the task of measuring snowfall from Chapter 10, say you have the following measurements (in inches) of snowfall:

7.8, 9.4, 10.0, 7.9, 9.4, 7.0, 7.0, 7.1, 8.9, 7.4

What is your 99.9 percent confidence interval for the true value of snowfall?
# Solution 2

```python
import numpy as np
from scipy.stats import norm

snowfall = np.array([7.8, 9.4, 10.0, 7.9, 9.4, 7.0, 7.0, 7.1, 8.9, 7.4])
print(snowfall)
mean = snowfall.mean()
std = snowfall.std()

dist = norm(loc = mean , scale = std)
print(f"Upper bound is {dist.ppf(.9995)} and the lower bound is {dist.ppf(0.0005)}")
```

# Problem 3
A child is going door to door selling candy bars. So far she has visited 30 houses and sold 10 candy bars. She will visit 40 more houses today. What is the 95 percent confidence interval for how many candy bars she will sell the rest of the day?
# Solution 3

```python
from scipy.stats import beta
houses_visited =30
houses_sold = 10
houses_not_sold = houses_visited-houses_sold
houses_to_visit = 40

dist = beta(houses_sold,houses_not_sold)
high_sales_rate = dist.ppf(.975)
low_sales_rate = dist.ppf(.025)
print(f"The top range of the likely sales range is {high_sales_rate} and the low range is {low_sales_rate}")
print(f"The top range of likely sales in the next {houses_to_visit} today is {round((high_sales_rate*houses_to_visit))} and the low range is {round((low_sales_rate*houses_to_visit))}")
```
